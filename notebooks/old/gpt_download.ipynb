```python
!pip install -q git+https://github.com/huggingface/transformers.git
!pip install -q accelerate

import os
import torch
from transformers import GPTJForCausalLM, AutoTokenizer

device_name = 'cuda' if torch.cuda.is_available() else 'cpu'
if device_name != 'cuda':
  raise Exception('Not using GPU')

device = torch.device(device_name)

path_model = PATH_DATA + 'gpt-j.pt'
path_tokenizer = PATH_DATA + 'gpt-j-tok.pt'

if not os.path.exists(path_model):
  model = GPTJForCausalLM.from_pretrained('EleutherAI/gpt-j-6B', revision='float16', low_cpu_mem_usage=True)
  torch.save(model, path_model)

if not os.path.exists(path_tokenizer):
  tokenizer = AutoTokenizer.from_pretrained('EleutherAI/gpt-j-6B')
  torch.save(tokenizer, path_tokenizer)
  
```
