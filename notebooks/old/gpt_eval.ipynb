```python
!pip install -q git+https://github.com/huggingface/transformers.git
!pip install -q accelerate

import os
import torch
from transformers import GPTJForCausalLM, AutoTokenizer

device_name = 'cuda' if torch.cuda.is_available() else 'cpu'
if device_name != 'cuda':
  raise Exception('Not using GPU')

device = torch.device(device_name)

# model
path_model = PATH_DATA + 'gpt-j.pt'
model = torch.load(path_model)
model.to(device)

# tokenizer
path_tokenizer = PATH_DATA + 'gpt-j-tok.pt'
tokenizer = torch.load(path_tokenizer)
  
```

```python
prompt_1 = 'The key benefit of TF-IDF are'

prompt_2 = '''
mov [ebp+var_A], dx
mov edx, [ebp+var_E]
mov eax, [ebp+arg_4]
movzx ecx, word ptr [eax+edx*2]
push ecx
'''

prompt_3 = '''
mov ebp var_A dx
mov edx ebp var_E
mov eax ebp arg_4
movzx ecx word ptr eax edx 2
push ecx
'''

# with open('0A32eTdBKayjCWhZqDOQ.asm', 'r', encoding='utf-8', errors='ignore') as f:
#     content = f.read()
#     cleaned = ''.join(content.split(' '))
#     prompt_3 = cleaned[:2048]
# print(prompt_3)

prompt = prompt_3
input = tokenizer(prompt, return_tensors="pt")
input.to(device)
```

```python
gen_tokens = model.generate(
    input.input_ids,
    do_sample=True,
    temperature=0.9,
    # max_length=100,
    max_new_tokens=100
)

tokenizer.batch_decode(gen_tokens)
gen_text = tokenizer.batch_decode(gen_tokens)[0]

new_text = gen_text.replace(prompt, '')

print('---')
print(prompt)
print('---')
print(new_text)
```
