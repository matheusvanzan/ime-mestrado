# Metrics Multi

```python
import evaluate
import seaborn as sn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
import drive.Shareddrives.GPTJ.project.metrics as metrics


def format(n):
    n = round(n, 4)
    return str(n).replace('.', ',')


def results(p):
    try:    
        df = pd.read_csv(PATH_DATA + p)
        predictions = list(df['pred_all'])
        references = list(df['ref'])
        # print('predictions', predictions)
        # print('references', references)
        print('len(references)', len(references))
        m = metrics.get_metrics(references, predictions)
        print('good', m['good'])
        print('bad', m['bad'])
        print('matrix', m['matrix'])
        print('acc', 'f1')
        print(format(m['accuracy']) + ';' + ';'.join([format(x) for x in m['f1']]))
    except:
        print('not done')

# matrix = m['matrix']
# matrix_100 = 100* matrix / matrix.astype(np.float).sum(axis=1)
# print(m)
# plt.figure(figsize = (10,7))
# sn.heatmap(matrix, annot=True, fmt='g')
```

```python
# 102400 - 1-fold
paths_1 = [f'/gpt2/all.limit-102400.chunk-32.epochs-{i}.batch-160/results.csv' for i in range(1, 5+1)]

for p in paths_1:
    print(p)
    results(p)    
    print('---')
```

```python
# 10240 - 10-fold
paths_2 = [f'/gpt2/all.limit-10240.fold-{i}.chunk-32.epochs-1.batch-160/results.csv' for i in range(1, 10+1)]

for p in paths_2:
    print(p)
    results(p)    
    print('---')
```

```python
# 102400 - 10-fold
paths_3 = [f'/gpt2/all.limit-102400.fold-{i}.chunk-32.epochs-1.batch-160/results.csv' for i in range(1, 10+1)]

for p in paths_3:
    print(p)
    results(p)    
    print('---')
```
