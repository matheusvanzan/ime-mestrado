```python
from matplotlib import pyplot as plt
from scipy.interpolate import make_interp_spline, BSpline
import numpy as np

import matplotlib.colors as mcolors

colors = [c for _, c in mcolors.TABLEAU_COLORS.items()]

def slope(x1, y1, x2, y2):
    if x1 == x2: return '0'
    slope = (y2-y1)/(x2-x1)
    slope_round = round(slope, 4)
    slope_str = str(slope_round).replace('.', ',')
    return slope_str

def plot_df(exp, df, var_x, var_y):
    defaults = {'epochs': 100, 'layer_dim': 256, 'n_layers': 1}
    vars_ = [k for k in defaults]
    vars_.remove(var_x[0])
    
    for v in vars_:
        df = df.loc[df[v] == defaults[v]]
        # df.drop(v, axis=1) # 1 for columns
        
    plt.figure()
    # plt.plot(df[var_x[0]], df[var_x[0]], label='id', color=colors[0])
    
    optmizers = ['SGD', 'RMSprop', 'Adam', 'Adadelta', 'Adagrad', 'Adamax', 'Nadam']
    print(f'E{exp}: {var_x[1]} x {var_y[1]}')
    
    for i, o in enumerate(optmizers):
        df_tmp = df.loc[df['optmizer'] == o]
        df_tmp = df_tmp.sort_values(var_x[0])
        
        # print(df_tmp)
        
        # total slope
        min_x = df_tmp[var_x[0]].min()
        max_x = df_tmp[var_x[0]].max()
        min_y = df_tmp[var_y[0]].min()
        max_y = df_tmp[var_y[0]].max()
        slope_str = slope(min_x, min_y, max_x, max_y)
        
        # step slope
        last_x = None
        last_y = None
        slopes_aux = []
        for row in df_tmp.iterrows():
            current_x = row[1][var_x[0]]
            current_y = row[1][var_y[0]]
            if last_x and last_y:
                slope_str_aux = slope(current_x, current_y, last_x, last_y)
                slopes_aux.append(slope_str_aux)
            last_x = current_x
            last_y = current_y
            
        # print slopes
        print(f'  {i} - {o} - slope: {slope_str} - slope_aux: {slopes_aux}')
        
        # plot charts
        plt.plot(df_tmp[var_x[0]], df_tmp[var_y[0]], label=o, color=colors[i])
        plt.xlabel(var_x[1])
        plt.ylabel(var_y[1])
        plt.axis('equal')
        params = {'mathtext.default': 'regular' }          
        plt.rcParams.update(params)
        plt.title(f'$E_{exp}$: {var_x[1]} x {var_y[1]}')
        plt.legend(loc='upper left')
        plt.gca()
        
    path_img = f'/home/jovyan/data/img/paa_{var_x[0]}_{var_y[0]}.png'
    plt.savefig(path_img, dpi=300, facecolor=(1.0, 1.0, 1.0, 1.0))
        
    return df 

exp = 0
for f1 in [('layer_dim', 'Neurônios (N)'), ('n_layers', 'Camadas (M)'), ('epochs', 'Épocas (E)')]:
    for f2 in [('time_elapsed', 'Tempo (min)'), ('ptime_elapsed', 'PTempo (min)'), ('mem_max', 'Memoria (MB)')]:
        exp += 1
        df_new = plot_df(exp, df, f1, f2)

# df_new
```
