```python
import pandas as pd
from matplotlib import pyplot as plt

path_csv = '/home/jovyan/data'
path_img = '/home/jovyan/code/img'

def xy_to_min(x, y):
    x = x/60
    x = x[::60]
    y = y[::60]
    return x, y

def plot_df(values, path_img, title, xlabel, ylabel):
    for value in values:
        path_csv, label, color, column = value
        df = pd.read_csv(path_csv, index_col = 0)
        
        print(df.shape)
        
        df = df[df[column] > 0]
        y = df[column].values
        
        x, y = xy_to_min(x, y)

        plt.plot(df.index, y, label=label, color=color)
        
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.legend(loc='upper left')
    plt.gca()
    plt.savefig(path_img, dpi=300, facecolor=(1.0, 1.0, 1.0, 1.0))
    
```

```python
'''
b: blue
g: green
r: red
c: cyan
m: magenta
y: yellow
k: black
w: white
'''

values = [
    (f'{path_csv}/mem_g1_ms2_mf1000.csv', 'Teste 1', 'b', 'mem'),
    (f'{path_csv}/mem_g2_ms2_mf1000.csv', 'Teste 2', 'r', 'mem'),
]

plot_df(
    values,
    path_img=f'{path_img}/mem_tfidf_1.png',
    title='Custo de memória RAM TfidfVectorizer',
    xlabel='Tempo (min)',
    ylabel='Memória RAM (GB)'
)
```
