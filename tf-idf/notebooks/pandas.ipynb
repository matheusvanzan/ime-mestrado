```python
from collections import Counter
import functools
import operator 

my_dict = [Counter({'a': 5, 'b': 1, 'c': 2}),
           Counter({'a': 2, 'b': 5}),
           Counter({'a': 10, 'c': 10})]

# sum the values with same keys
res = functools.reduce(operator.add, my_dict)

print("New dict : ", res)
len(res)
```
